```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Sample data: sq. feet (X) and house price in thousands (y)
X = np.array([1300, 1500, 1700, 1900, 2100, 2500, 3600, 6900]).reshape(-1, 1)
y = np.array([240, 310, 330, 360, 400 , 420 , 460, 800
])

# Create and train the model
model = LinearRegression()
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Print model parameters
print(f"Intercept (b): {model.intercept_:.2f}")
print(f"Coefficient (m): {model.coef_[0]:.2f}")

# Evaluate the model
print(f"Mean Squared Error: {mean_squared_error(y, y_pred):.2f}")
print(f"R² Score: {r2_score(y, y_pred):.2f}")

# Plot the results
plt.scatter(X, y, color='blue', label='Actual')
plt.plot(X, y_pred, color='red', label='Predicted')
plt.xlabel("Square Footage")
plt.ylabel("House Price (in $1000s)")
plt.title("Linear Regression: House Price Prediction")
plt.legend()
plt.show()
```
